```python
# 1. Simple Linear Regression python code.
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
#1. Prepare the data
# Independent variable (features) - must be a 2D array for scikit-learn
X = np.array([500, 1000, 1500, 1750, 2100]).reshape(-1, 1)
# Dependent variable (target)
y = np.array([1, 2, 2.5, 3.2, 3.9])

#2. Create and train the model
model = LinearRegression()
model.fit(X, y)

#3. Get the model's coefficients
intercept = model.intercept_
coefficient = model.coef_[0]

print(f"Intercept (b0): {intercept}")
print(f"Coefficient (b1): {coefficient}")

#4. Make predictions
y_pred = model.predict(X)

#5. Visualize the results (optional)
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (y)")
plt.title("Simple Linear Regression")
plt.legend()
plt.grid(True)
plt.show()

#6 Prediction of the new data
new_data =  np.array([2500, 3000]).reshape(-1, 1)
predicted_value = model.predict(new_data)
print(f"Prediction for new data {new_data}: {predicted_value}")

#7- Now place the predicted value in original data
X=np.append(X,new_data).reshape(-1,1)
y=np.append(y,predicted_value)

#8- Plot a scatter plot with new value
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (y)")
plt.title("Simple Linear Regression")
plt.legend()
plt.grid(True)
plt.show()

```
