1-	Central Tendencies


```python
import pandas as pd
from sklearn.impute import SimpleImputer

df = pd.read_csv('https://raw.githubusercontent.com/venky14/Machine-Learning-with-Iris-Dataset/refs/heads/master/Iris.csv')
print(df)

df2 = df.drop(index=range(10)).reset_index(drop=True)
df2.iloc[0:5, 1] = 0
df_copy = df2.copy()

num_cols = df2.select_dtypes(include='number')
cat_cols = df2.select_dtypes(exclude='number')

imp = SimpleImputer(missing_values=0, strategy='mean')
num_imputed = pd.DataFrame(imp.fit_transform(num_cols), columns=num_cols.columns)

df_imputed = pd.concat([num_imputed, cat_cols.reset_index(drop=True)], axis=1)

mean_val = num_imputed.mean()
median_val= num_imputed.median()
mode_val = num_imputed.mode().iloc[0]

diff = num_imputed - num_cols

print("\nCentral Tendency")
print("Mean:\n", mean_val)
print("Median:\n", median_val)
print("Mode:\n", mode_val)
print("\nDifference:\n", diff.head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Iris dataset
iris = pd.read_csv('https://raw.githubusercontent.com/venky14/Machine-Learning-with-Iris-Dataset/refs/heads/master/Iris.csv')

# 1. Bar plot - count of each species
species_counts = iris['Species'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(species_counts.index, species_counts.values, color=['red', 'green', 'blue'])
plt.title("Count of Each Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()

# 2. Histogram - SepalLengthCm distribution
plt.figure(figsize=(6,4))
plt.hist(iris['SepalLengthCm'], bins=15, color='skyblue', edgecolor='black')
plt.title("Sepal Length Distribution")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()

# 3. Pie chart - proportion of species
plt.figure(figsize=(6,6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%',
        colors=['#ff9999','#66b3ff','#99ff99'])
plt.title("Species Proportion")
plt.show()

# 4. Scatter plot - SepalLengthCm vs PetalLengthCm
plt.figure(figsize=(6,4))
for species in iris['Species'].unique():
    subset = iris[iris['Species'] == species]
    plt.scatter(subset['SepalLengthCm'], subset['PetalLengthCm'], label=species)
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend()
plt.show()

# 5. Box plot - PetalWidthCm by Species
data = [iris[iris['Species'] == sp]['PetalWidthCm'] for sp in iris['Species'].unique()]
plt.figure(figsize=(6,4))
plt.boxplot(data, labels=iris['Species'].unique())
plt.title("Petal Width by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.show()
```
